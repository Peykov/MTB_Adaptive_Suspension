# Adpative Suspension Mountain Bike
#### Real-time terrain analysis and optimization for mountain bike suspension efficiency

### Problem:
- #### Not optimal response of the suspension to the variety of the terrain conditions, features, obstacles and rider's input.
- ####  So many variables left solely to the rider to adjust constantly with his physical input (body position, weight distribution) during the descent. The suspension responsiveness and efficiency stays constant through the ride. 

#### Possible big improvements that could be achieved by optimizing the suspension settings in real-time:

    - Turns ( high low speed compression, fast high speed rebound)
    - Jumps ( mid-stroke support, fast high speed rebound), preload 
    - Drops ( softer combination before the drop, firmer on the landing )
    - Rock gardens ( mid-stroke support, fast high speed rebound)
    - Roots ( fast high speed rebound, fast highspeed comression)
    - Straight steep lines ( soft overall combination of settings)
    - Flat corners etc.

    - Could be adapted for tyre pressure live optimization as well

### Output: Rebound and Compression settings
    - Rebound: How fast the suspension returns to its original position after being compressed 
        - fast speed rebound (range(1,12)); low speed rebound (range(1,12))
    - Compression: How much and how fast the suspension compresses when a force is applied to it ( firm - soft)
        - low speed compression (range(1:10)) - fast speed compression (range(1:10))
    
- The output has to be send as signal in time ( amplitude and frequency) converted in mechanical rotation of rebound and compression knobs of the suspension fork and shock.

### Input: Suspension travel, force applied and Trail(road) detection 
- Possible input recievers: 
    - Lidar
    - Camera
    - IMU ( inertial measurement unit)
